```python
table = pd.crosstab(dataset['Profession'],dataset['Depression'])
table
```

```python
from scipy.stats import chi2_contingency
```

```python
chi2, p, dof, ex = chi2_contingency(table)
print(f"Chi-square Test Result (p-value): chi2 = {chi2}, p-value = {p}")
```

```python
table = pd.crosstab(dataset['Gender'],dataset['Depression'])
table
chi2, p, dof, ex = chi2_contingency(table)
print(f"Chi-square Test Result (p-value): chi2 = {chi2}, p-value = {p}")

#Since the p-value is much greater than 0.05, we fail to reject the null hypothesis.
#This means there is no statistically significant association between Gender and Depression in the dataset.
```

```python
table = pd.crosstab(dataset['City'],dataset['Depression'])
table
chi2, p, dof, ex = chi2_contingency(table)
print(f"Chi-square Test Result (p-value): chi2 = {chi2}, p-value = {p}")


#A high Chi-Square statistic indicates a large difference between observed and expected frequencies.

#The p-value is extremely small, much lower than the typical significance level (Î±=0.05Î±=0.05).
#This means there is strong evidence to reject the null hypothesis.
```

```python
table = pd.crosstab(dataset['Sleep Duration'],dataset['Depression'])
table
chi2, p, dof, ex = chi2_contingency(table)
print(f"Chi-square Test Result (p-value): chi2 = {chi2}, p-value = {p}")

#p-value is extremely small (1.0653e-58), we reject the null hypothesis. 
#This means there is strong evidence to suggest that Sleep Duration and Depression are not independent and 
#that there is a significant association between them.
```
